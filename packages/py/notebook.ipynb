#### Step 1: Define your agent

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from langchain.messages import HumanMessage, AIMessage, ToolMessage

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.7,
    google_api_key="<your-google-api-key>",
)

def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

def add(a: int, b: int) -> int:
    """Add two numbers."""
    return a + b

def subtract(a: int, b: int) -> int:
    """Subtract two numbers."""
    return a - b

res = create_agent(llm, [multiply, add, subtract], system_prompt="You are a helpful assistant. Use the tools when needed.")
result = res.invoke({"messages": [HumanMessage("Give me the result of 2 * 3 + 4 - 1")]})
messages = result["messages"]

# messages = [HumanMessage(content='Give me the result of 2 * 3 + 4 - 1', additional_kwargs={}, response_metadata={}, id='8b13f812-54cd-449b-a621-2484f38996ba'),
#  AIMessage(content='', additional_kwargs={'function_call': {'name': 'multiply', 'arguments': '{"a": 2, "b": 3}'}}, response_metadata={'finish_reason': 'STOP', 'model_name': 'gemini-2.5-flash-lite', 'safety_ratings': [], 'model_provider': 'google_genai'}, id='lc_run--019caf9e-8e2b-78b0-b5c2-43015a340495-0', tool_calls=[{'name': 'multiply', 'args': {'a': 2, 'b': 3}, 'id': '796ad636-b6fa-47b7-9b7a-ae6f1ccb53f9', 'type': 'tool_call'},{'name': 'add', 'args': {'a': 2, 'b': 2}, 'id': '796ad636-b6fa-47b7-9b7a-ae6f1ccb53f9', 'type': 'tool_call'}], invalid_tool_calls=[], usage_metadata={'input_tokens': 164, 'output_tokens': 18, 'total_tokens': 182, 'input_token_details': {'cache_read': 0}}),
#  ToolMessage(content='6', name='multiply', id='6ee022d8-7344-41b2-92e6-cc58bc8512e8', tool_call_id='796ad636-b6fa-47b7-9b7a-ae6f1ccb53f9'),
#  ToolMessage(content='4', name='add', id='6ee022d8-7344-41b2-92e6-cc58bc8512e9', tool_call_id='796ad636-b6fa-47b7-9b7a-ae6f1ccb53f9'),
#  AIMessage(content='', additional_kwargs={'function_call': {'name': 'add', 'arguments': '{"a": 6, "b": 4}'}}, response_metadata={'finish_reason': 'STOP', 'model_name': 'gemini-2.5-flash-lite', 'safety_ratings': [], 'model_provider': 'google_genai'}, id='lc_run--019caf9e-92e5-7842-a7a3-2e64a0830469-0', tool_calls=[{'name': 'add', 'args': {'a': 6, 'b': 4}, 'id': 'ec640e0b-cb79-44dc-8928-adc140bee51e', 'type': 'tool_call'}], invalid_tool_calls=[], usage_metadata={'input_tokens': 195, 'output_tokens': 18, 'total_tokens': 213, 'input_token_details': {'cache_read': 0}}),
#  ToolMessage(content='10', name='add', id='7d0bc1e1-c51f-4593-bcbb-aa1b9064f9c2', tool_call_id='ec640e0b-cb79-44dc-8928-adc140bee51e'),
#  AIMessage(content='', additional_kwargs={'function_call': {'name': 'subtract', 'arguments': '{"a": 10, "b": 1}'}}, response_metadata={'finish_reason': 'STOP', 'model_name': 'gemini-2.5-flash-lite', 'safety_ratings': [], 'model_provider': 'google_genai'}, id='lc_run--019caf9e-95d0-7d11-a304-5069e70e1672-0', tool_calls=[{'name': 'subtract', 'args': {'a': 10, 'b': 1}, 'id': '604f9e33-db8f-4b87-930f-660bc4646ff0', 'type': 'tool_call'}], invalid_tool_calls=[], usage_metadata={'input_tokens': 227, 'output_tokens': 19, 'total_tokens': 246, 'input_token_details': {'cache_read': 0}}),
#  ToolMessage(content='9', name='subtract', id='2baefc5a-cebe-4c0a-9524-f9d036a80c38', tool_call_id='604f9e33-db8f-4b87-930f-660bc4646ff0'),
#  AIMessage(content='The result is 9.', additional_kwargs={}, response_metadata={'finish_reason': 'STOP', 'model_name': 'gemini-2.5-flash-lite', 'safety_ratings': [], 'model_provider': 'google_genai'}, id='lc_run--019caf9e-9969-73e3-a774-571d7429f142-0', tool_calls=[], invalid_tool_calls=[], usage_metadata={'input_tokens': 259, 'output_tokens': 6, 'total_tokens': 265, 'input_token_details': {'cache_read': 0}})]
```
