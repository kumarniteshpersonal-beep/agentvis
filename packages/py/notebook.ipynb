#### How to use **agentvis** ?

#### Step 1: Define your agent

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from langchain.messages import HumanMessage, AIMessage, ToolMessage

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.7,
    google_api_key="AIzaSyDgf7z9CkrZoSxUVWhrjIeRTSV6pxu2E8k",
)

def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

def add(a: int, b: int) -> int:
    """Add two numbers."""
    return a + b

def subtract(a: int, b: int) -> int:
    """Subtract two numbers."""
    return a - b

res = create_agent(llm, [multiply, add, subtract], system_prompt="You are a helpful assistant. Use the tools when needed.")
result = res.invoke({"messages": [HumanMessage("Give me the result of 2 * 3 + 4 - 1")]})
messages = result["messages"]
```

#### Step 2: Use **agentvis** to build reasoning tree

```python
from agentvis.framework.langchain import LangChainFrameWork
from agentvis.core.export import ExportFactory

frames = LangChainFrameWork().build_frames(messages)
export = ExportFactory.export_frames(frames=frames, export_strategy="link") # export_strategy = "link" | "json"
```
